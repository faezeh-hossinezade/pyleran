```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
n=200
```

```python
melons = pd.DataFrame({
    'length': np.random.normal(24, 3, n),
    'width': np.random.normal(0, 7, n),
    'weight' : np.random.normal(10, 4, n),
    'class': np.zeros(n, dtype='int')
})

ballon = pd.DataFrame({
    'length': np.random.normal(32, 3, n),
    'width': np.random.normal(80, 7, n),
    'weight' : np.random.normal(50, 4, n),
    'class': np.ones(n, dtype='int')}
)


shapes= pd.concat([melons, ballon])
shapes
```

```python
plt.rcParams['xtick.labelsize'] = 5
plt.rcParams['ytick.labelsize'] = 5
plt.rcParams['legend.fontsize'] = 5
plt.rcParams['axes.labelsize'] = 5
plt.rcParams['axes.titlesize'] = 5
```

```python
fig = plt.figure(figsize=(4,4))

ax = plt.axes(111,projection='3d')
data_class_0 = shapes[shapes['class'] == 0]
ax.scatter(data_class_0 ['length'], data_class_0 ['width'], data_class_0 ['weight'],label='Melons',alpha=0.5)
data_class_1 = shapes[shapes['class'] == 1]
ax.scatter(data_class_1 ['length'], data_class_1 ['width'], data_class_1 ['weight'],marker='^',label='Ballons',alpha=0.5,c='orange')

ax.set_zlabel('Z Label', rotation=45) 
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label',  labelpad=-4)
plt.show()
```
